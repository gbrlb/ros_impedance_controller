```python
from xml.etree import ElementTree as ET
from xml.etree import ElementTree as ET

# Load the XML
tree = ET.parse('joint_torques.xml')
root = tree.getroot()
print(root.tag)


for i in range(3):
    for j in range(4):
        field = root.findall(f"./table/plots/row_{i}/column_{j}//field")
        value = i + 3 * j
        for f in field:
            # only include fields that have the string 'effort' in them
            if 'effort' in f.text:
                f.text = f.text.rsplit('/', 1)[0] + '/' + str(value)

        scales_y = root.findall(f"./table/plots/row_{i}/column_{j}//y_axis/scale")
        # replace the scale/absolute_maximum element to 10:
        for scale_y in scales_y:
            for element in scale_y:
                # print(element.tag, element.text)
                if element.tag == 'absolute_maximum':
                    element.text = '10'
                if element.tag == 'absolute_minimum':
                    element.text = '-10'
                if element.tag == 'type':
                    element.text = '1'

        title = root.find(f"./table/plots/row_{i}/column_{j}/title")
        labels_quadruped = ["LF_HAA", "LF_HFE","LF_KFE","LH_HAA", "LH_HFE","LH_KFE","RF_HAA", "RF_HFE","RF_KFE","RH_HAA", "RH_HFE","RH_KFE"]
        title.text = labels_quadruped[value]
        # print(title.text)

        datas  = root.findall(f"./table/plots/row_{i}/column_{j}//data")

        for data in datas:
            for element in data:
                if element.tag == 'circular_buffer_capacity':
                    element.text = '500'
                if element.tag == 'type':
                    element.text = '2'
                print(element.tag, element.text)

tree.write('joint_torques.xml')
```
